**Dataset Used:** Maternal Health Risk Data Set

## 1. Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Reading Data

```python
df = pd.read_csv("Maternal Health Risk Data Set.csv")
```

## 3. Exploring Data

```python
# Display first few rows
print("First five rows of the dataset:")
display(df.head())

# Shape of the dataset
print("\nShape of the dataset:")
print(df.shape)

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Data Types
print("\nData types of each column:")
print(df.dtypes)

# Unique Values
print("\nNumber of unique values in each column:")
print(df.nunique())

# Summary Statistics
print("\nSummary Statistics:")
print(df.describe())
```

## 4. Removing Duplicates

```python
df = df.drop_duplicates().reset_index(drop=True)
```

## 5. Data Visualization

```python
# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Age'], y=df['HeartRate'], hue=df['RiskLevel'])
plt.title("Scatter Plot of Age vs. Heart Rate")
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8,5))
df['RiskLevel'].value_counts().plot(kind='bar', color=['blue', 'orange', 'red'])
plt.title("Bar Plot of Risk Level Distribution")
plt.xlabel("Risk Level")
plt.ylabel("Count")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(x=df['RiskLevel'], y=df['BodyTemp'])
plt.title("Box Plot of Body Temperature by Risk Level")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(6,6))
df['RiskLevel'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'orange', 'red'])
plt.title("Pie Chart of Risk Levels")
plt.ylabel("")
plt.show()
```

```python
# Line Chart
plt.figure(figsize=(8,5))
plt.plot(df['Age'].sort_values().values, df['HeartRate'].sort_values().values, marker='o', linestyle='-')
plt.title("Line Chart of Age vs. Heart Rate")
plt.xlabel("Age")
plt.ylabel("Heart Rate")
plt.show()
```
